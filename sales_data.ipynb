# Import Libraries

```python
# To get the data
from bs4 import BeautifulSoup
import requests

# For text preprocessing
import pandas as pd
import re
```

# Getting the data

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
```

```python
base_site = 'https://en.wikipedia.org/wiki/Pok%C3%A9mon_(video_game_series)#Reception'
response = requests.get(base_site, headers = headers)
response
```

# Parsing the Data

```python
soup = BeautifulSoup(response.content, 'lxml')
```

```python
results = soup.find_all(['table'], class_='wikitable plainrowheaders')
```

```python
# https://medium.com/analytics-vidhya/web-scraping-a-wikipedia-table-into-a-dataframe-c52617e1f451
df = pd.DataFrame(pd.read_html(str(results))[0])
df
```

## Text Preprocessing

```python
def remove_square_brackets(text):
    text = re.sub(r'\[.*?\]', '', text)
    return text
```

```python
# Removing square brackets and their content
# https://stackoverflow.com/questions/42324466/python-regular-expression-to-remove-all-square-brackets-and-their-contents
df['Units sold(in millions)'] = df['Units sold(in millions)'].apply(remove_square_brackets)
df['GameRankings'] = df['GameRankings'].apply(remove_square_brackets)
df['Metacritic'] = df['Metacritic'].apply(remove_square_brackets)
df.head()
```

```python
df.to_csv('games_sales.csv', index=False)
```
